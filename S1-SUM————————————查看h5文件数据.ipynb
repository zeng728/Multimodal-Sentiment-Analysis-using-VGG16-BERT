```python
import h5py

# 打开h5py文件
with h5py.File('MVSA_Single/data_train.h5', 'r') as f:
    
    # 遍历所有数据集
    for dset_name in f.keys():
        print('Dataset:', dset_name)
        
        # 获取当前数据集
        dset = f[dset_name]
        
        # 打印数据集的shape
        data_shape = dset.shape
        print('Data Shape:', data_shape)        
        
        # 打印数据集的前10行
        data = dset[:]
        print('Data:')
        for i, row in enumerate(data[:1]):
            print(row)
```

```python
text = []
with h5py.File('MVSA_Single/data_train.h5', 'r') as f:
    for item in f['texts']:
        text.append(item.decode('utf-8'))
    print(text)
```
